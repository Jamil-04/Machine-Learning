```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Step 1: Load dataset
# Upload file Position_Salaries.csv ke Google Colab terlebih dahulu
from google.colab import files
uploaded = files.upload()

# Membaca dataset
df = pd.read_csv('Position_Salaries.csv')

# Menampilkan 5 baris pertama dataset
df.head()

# Kesimpulan Step 1:
# Dataset Position_Salaries.csv berhasil di-load. Dataset ini terdiri dari dua fitur: "Level" dan "Salary".
```

```python
# Step 2: Memisahkan fitur dan target
X = df.iloc[:, 1:-1].values  # Kolom "Level"
y = df.iloc[:, -1].values    # Kolom "Salary"

# Kesimpulan Step 2:
# Fitur (Level) dan target (Salary) telah dipisahkan.
```

```python
# Step 3: Membuat model Linear Regression (untuk perbandingan)
linear_regressor = LinearRegression()
linear_regressor.fit(X, y)

# Kesimpulan Step 3:
# Model Linear Regression sederhana telah dibuat untuk perbandingan hasil dengan Polynomial Regression.
```

```python
# Step 4: Membuat model Polynomial Regression
# Membuat polynomial features (derajat 4)
poly_features = PolynomialFeatures(degree=4)
X_poly = poly_features.fit_transform(X)

# Membuat model Polynomial Regression dengan fitur polynomial
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y)

# Kesimpulan Step 4:
# Model Polynomial Regression dengan derajat 4 telah dibuat.
```

```python
# Step 5: Visualisasi hasil model Linear Regression
plt.scatter(X, y, color='red')  # Titik data asli
plt.plot(X, linear_regressor.predict(X), color='blue')  # Garis regresi linear
plt.title('Linear Regression (Position vs Salary)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

# Kesimpulan Step 5:
# Grafik ini menunjukkan prediksi garis linear yang dibuat oleh model Linear Regression sederhana.
# Garis linear ini tidak cocok untuk data yang jelas non-linear.
```

```python
# Step 6: Visualisasi hasil model Polynomial Regression
# Membuat kurva polynomial yang lebih halus
X_grid = np.arange(min(X), max(X), 0.1)  # Membuat grid nilai untuk plotting yang lebih halus
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color='red')  # Titik data asli
plt.plot(X_grid, poly_regressor.predict(poly_features.transform(X_grid)), color='blue')  # Garis polynomial
plt.title('Polynomial Regression (Position vs Salary)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

# Kesimpulan Step 6:
# Grafik ini menunjukkan kurva polynomial yang lebih sesuai dengan data non-linear. Kurva mengikuti tren kenaikan gaji yang lebih realistis dengan peningkatan level posisi.
```

```python
# Step 7: Membuat prediksi dengan Polynomial Regression
# Memprediksi gaji untuk level tertentu menggunakan Polynomial Regression
level = 6.5
predicted_salary_poly = poly_regressor.predict(poly_features.transform([[level]]))
print("Prediksi Gaji untuk level {}: {:.2f}".format(level, predicted_salary_poly[0]))

# Kesimpulan Step 7:
# Model Polynomial Regression dapat digunakan untuk memprediksi gaji pada level tertentu. Misalnya, pada level 6.5, model memprediksi gaji tertentu.
```
