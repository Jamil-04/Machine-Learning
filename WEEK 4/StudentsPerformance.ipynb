Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

```python
# Fungsi untuk menampilkan pie chart untuk kolom bertipe kategorikal
def plot_categorical_distribution(df, column_name):
    # Menghitung frekuensi dan persentase setiap kategori
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Membuat pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribusi {column_name}')
    plt.axis('equal')  # Membuat chart berbentuk lingkaran
    plt.show()

# Menampilkan pie chart untuk kolom bertipe kategorikal
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

for col in categorical_columns:
    plot_categorical_distribution(df, col)
```

```python
# Fungsi untuk menampilkan pie chart dengan efek 3D dan warna pastel
def plot_pie_chart(column_name):
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Menggunakan palet warna pastel dari 'Pastel1'
    colors = plt.cm.Pastel1.colors

    plt.figure(figsize=(8, 8))
    # Membuat efek bayangan dengan `shadow=True` dan `explode` untuk sedikit menggeser setiap bagian
    explode = [0.05] * len(labels)  # Memberikan efek terpisah sedikit pada setiap kategori
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors[:len(labels)],
            shadow=True, explode=explode)

    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}', fontsize=14, fontweight='bold', color='#333333')
    plt.show()

# Plot pie chart dengan warna pastel dan efek 3D
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
```

```python
def plot_pie_chart_3d(column_name):
    # Menghitung jumlah masing-masing kategori
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Membuat pie chart dengan warna pastel dan 3D effect
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))
    pastel_colors = sns.color_palette('pastel')

    # Membuat pie chart dengan shadow dan efek explode
    wedges, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%',
                                      startangle=90, colors=pastel_colors, shadow=True,
                                      explode=[0.05]*len(labels))  # Explode sedikit semua bagian

    # Membuat teks lebih jelas
    for text in autotexts:
        text.set_color('black')

    # Menambahkan efek shadow dan properti 3D
    plt.setp(wedges, edgecolor='gray')

    plt.title(f'Percentage of {column_name} (3D View)')
    plt.show()

# Panggil fungsi untuk berbagai kolom
plot_pie_chart_3d('gender')
plot_pie_chart_3d('race/ethnicity')
plot_pie_chart_3d('lunch')
plot_pie_chart_3d('parental level of education')
plot_pie_chart_3d('test preparation course')
```

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

```python
# Buat kolom baru untuk rata-rata nilai semua mata pelajaran
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Mengatur style visualisasi
sns.set(style="whitegrid")

# Ukuran figure untuk plot
plt.figure(figsize=(16, 10))

# Subplot 1: Rata-rata nilai berdasarkan Gender
plt.subplot(2, 3, 1)
sns.boxplot(x='gender', y='average score', data=df, palette='pastel')
plt.title('Average Score by Gender')

# Subplot 2: Rata-rata nilai berdasarkan Race/Ethnicity
plt.subplot(2, 3, 2)
sns.boxplot(x='race/ethnicity', y='average score', data=df, palette='pastel')
plt.title('Average Score by Race/Ethnicity')

# Subplot 3: Rata-rata nilai berdasarkan Parental Level of Education
plt.subplot(2, 3, 3)
sns.boxplot(x='parental level of education', y='average score', data=df, palette='pastel')
plt.xticks(rotation=45)  # Rotasi x-label agar tidak bertumpuk
plt.title('Average Score by Parental Education')

# Subplot 4: Rata-rata nilai berdasarkan Lunch
plt.subplot(2, 3, 4)
sns.boxplot(x='lunch', y='average score', data=df, palette='pastel')
plt.title('Average Score by Lunch')

# Subplot 5: Rata-rata nilai berdasarkan Test Preparation Course
plt.subplot(2, 3, 5)
sns.boxplot(x='test preparation course', y='average score', data=df, palette='pastel')
plt.title('Average Score by Test Preparation Course')

# Menampilkan plot
plt.tight_layout()
plt.show()

# Korelasi dengan kategori yang numerik (jika ada)
# Jika kategori dalam data numerik bisa dikonversi, berikut contohnya:
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Menggunakan dummy variables untuk kategorikal data agar dapat dihitung korelasinya
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Menampilkan korelasi
correlation_matrix = df_encoded.corr()
print(correlation_matrix['average score'].sort_values(ascending=False))
```
